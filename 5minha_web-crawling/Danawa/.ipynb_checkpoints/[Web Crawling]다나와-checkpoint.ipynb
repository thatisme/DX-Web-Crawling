### 다나와 URL 수집

#### 1) 다나와 '음식물처리기' 검색 결과 창 켜기

```python
driver = webdriver.Chrome()
url = 'https://prod.danawa.com/list/?cate=1023084'
driver.get(url)
driver.maximize_window()
```

#### 2) 품목 > 음식물 처리기 체크

```python
btn = driver.find_element(By.XPATH,'//*[@id="searchAttributeValue147493"]')
btn.click()
time.sleep(1)
```

#### 3) 처리방식 > 미생물 & 습식분쇄 + 미생물 체크

```python
btn = driver.find_element(By.XPATH,'//*[@id="searchAttributeValue782359"]')
btn.click()

btn = driver.find_element(By.XPATH,'//*[@id="searchAttributeValue782311"]')
btn.click()
time.sleep(1)
```

#### 4) 상품평 많은순 클릭

```python
btn = driver.find_element(By.XPATH,'//*[@id="productListArea"]/div[2]/div[1]/ul/li[5]/a')
btn.click()
time.sleep(1)
```

#### 5) 제품 href 추출

```python
# 하나의 페이지(30개)에 대하여 href를 추출하는 함수
def page_extract_href(list_href, list_name):   
    soup = bs(driver.page_source,'lxml')
    # 상품 정보 수집
    products = soup.find_all('div',attrs={'class': 'prod_info'})
    
    for p in products:
        # 상품 리뷰 링크 수집
        name = p.find('a',attrs={'name': 'productName'}).text.strip()
        review = p.find('div',attrs={'class': 'meta_item mt_comment'})
        if review:
            list_href.append(('https://prod.danawa.com/' + review.find('a')['href']).replace('N#','Y#'))
            list_name.append(name)
    return list_href, list_name
```

```python
list_href = []
list_name = []
df_href = pd.DataFrame(columns=['product_name','href'])

for i in tqdm(range(1,5,1)):
    btn = driver.find_element(By.XPATH,f'//*[@id="productListArea"]/div[4]/div/div/a[{i}]')
    btn.click()
    time.sleep(2)
    
    list_href, list_name = page_extract_href(list_href, list_name)
```

#### 6) 수집한 데이터들 저장

```python
df_href['product_name'] = list_name
df_href['href'] = list_href
df_href.set_index('product_name', inplace=True)
df_href.to_csv('./DanawaURL.csv')
```

```python
driver.close()
```

### 다나와 리뷰 수집

#### 1) href 불러오기

```python
load_danawa_href_pd = pd.read_csv('./DanawaURL.csv')
load_danawa_href_pd.set_index('product_name', inplace=True)
load_danawa_href_pd
```

#### 2) 제품별 리뷰 데이터 수집
- 제품별 페이지 접근(href) > 쇼핑몰 상품리뷰 데이터 수집 > 끝페이지까지 이동 후 처음으로 돌아가 반복
- 리뷰 제목
- 리뷰 점수
- 리뷰 날짜
- 리뷰 내용

```python
product_name_list,title_list, star_list, date_list, content_list = [], [], [], [], []
order_number = 1

for product_name, href in zip(load_danawa_href_pd.index, load_danawa_href_pd.href):
    driver = webdriver.Chrome()
    driver.get(href)
    driver.maximize_window()
    
    soup = bs(driver.page_source)
    reviews = soup.select('ul.rvw_list li.danawa-prodBlog-companyReview-clazz-more')

    try:
        while True:
            time.sleep(1)
            for review in reviews:
                product_name_list.append(product_name)
                title_list.append(review.select_one('.tit').text)
                star_list.append(review.select_one('.star_mask').text)
                date_list.append(review.select_one('.date').text)
                content_list.append(review.select_one('.atc').text)
            
            order_number += 1
            order = driver.find_element(By.CSS_SELECTOR, f'[data-pagenumber="{order_number}"]')
            order.click()
            time.sleep(1)
    except:
        order_number = 1
        continue
```

#### 3) 수집 데이터 저장

```python
danawa_review_data_pd = {'제품': product_name_list, '리뷰 제목': title_list, '리뷰 점수': star_list, '리뷰 날짜': date_list, '리뷰 내용': content_list}
danawa_review_data_pd = pd.DataFrame(danawa_review_data_pd)
danawa_review_data_pd
```

```python
danawa_review_data_pd.to_csv('./Danawa_review_data.csv')
```
